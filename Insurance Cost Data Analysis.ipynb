```python
#Import csv library to work with the data file
import csv
from decimal import Decimal 

#Create an empty for list for each variable in the dataset
ages = []
sex = []
bmi = []
children = []
smoker = []
region = []
charges = []

#Function to append row data to appropriate list
def load_list_data(lst , csv_file , column_name):
    with open('insurance.csv') as insurance_csv:
        data_reader = csv.DictReader(insurance_csv)
        for row in data_reader:
            lst.append(row[column_name])
        return lst
    
load_list_data(ages , 'insurance_csv' , 'age')
load_list_data(sex , 'insurance_csv' , 'sex')
load_list_data(bmi , 'insurance_csv' , 'bmi')
load_list_data(children , 'insurance_csv' , 'children')
load_list_data(smoker , 'insurance_csv' , 'smoker')
load_list_data(region , 'insurance_csv' , 'region')
load_list_data(charges , 'insurance_csv' , 'charges')
    
#Class for Patient Info
class PatientInfo:
    def __init__(self , age , sex , bmi , children , smoker , region , charges):
        self.age = age
        self.sex = sex
        self.bmi = bmi
        self.children = children
        self.smoker = smoker
        self.region = region
        self.charges = charges
    
    #Finds average age of patient group
    def average_age(self):
        total_age = 0
        for age in self.age:
            total_age = total_age + int(age)
        return total_age / len(ages)
    
    #Find which region has the most people and how many people it has
    def region_count(self):
        northwest = 0
        northeast = 0
        southeast = 0
        southwest = 0
        for region in self.region:
            if region == 'northwest':
                northwest += 1
            if region == 'northeast':
                northeast += 1
            if region == 'southeast':
                southeast += 1
            if region == 'southwest':
                southwest += 1
        if northwest > northeast and northwest > southeast and northwest > southwest:
            return ['NorthWest' , northwest]
        if northeast > northwest and northeast > southeast and northeast > southwest:
            return ['NorthEast' , northeast]
        if southeast > northwest and southeast > northeast and southeast > southwest:
            return ['SouthEast' , southeast]
        if southwest > northwest and southwest > northeast and southwest > southeast:
            return ['SouthWest' , southwest]
    
    #Finds the averge charge for smokers and non-smokers
    def smoker_difference(self):
        total_smoker = 0
        total_smoker_charge = 0
        total_non_smoker = 0
        total_non_smoker_charge = 0
        index = 0
        for smoker in self.smoker:
            if smoker == 'yes':
                total_smoker += 1
                total_smoker_charge += float(self.charges[index])
                index += 1
            else:
                total_non_smoker += 1
                total_non_smoker_charge += float(self.charges[index])
                index += 1
        return [(total_non_smoker_charge / total_non_smoker) , (total_smoker_charge / total_smoker)]
    
    #Finds the average age of parents
    def parent_average_age(self):
        total_parents = 0
        total_parents_cost = 0
        index = 0
        for child in self.children:
            if int(child) >= 1:
                total_parents += 1
                total_parents_cost += float(self.age[index])
                index += 1
            else:
                continue
        return total_parents_cost / total_parents

#Creates object for patients based on lists from data
patients = PatientInfo(ages , sex , bmi , children , smoker , region , charges)

#Creates a dictionary of key variables from calculations
key_variables = {'Average Age' : patients.average_age() , "Most Region" : patients.region_count() , "Smoker Charge" : patients.smoker_difference() , "Average Parent Age" : patients.parent_average_age()}

#Print statements on the key variables
print('The average age of patients is ' + str(patients.average_age()))
print(str(patients.region_count()[0]) + ' is the most populated region with ' + str(patients.region_count()[1]) + ' people')
print('Non-smoker\'s have an average charge of ' + str(patients.smoker_difference()[0]) + 'while smoker\'s have an average cost of ' +str(patients.smoker_difference()[1]))
print('The average age of parents in this group is ' + str(patients.parent_average_age()))
        
    
```
