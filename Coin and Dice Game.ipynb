I want to be able to guess the outcome of a random coin flip (heads/tails)

Create function that takes in a guess of user

```python
import random
```

```python
def coin_flip_game():
    
    #Ask if people want to play
    start_game = input ('Would you like to start the game Y/N: ')
    if start_game == "Y":
        
        #Set starting values for correct and total guesses
        index = 0
        correct_guesses = 0
        total_guesses = 0
        
        #Create loop of gameplay
        while index == 0:
            
            #Get the guess of the player
            coin_flip_guess = input ('Type your guess of heads or tails: ')
            print("Your guess is " + coin_flip_guess)
            
            #Flip the coin
            result = random.random()
            if result < 0.5:
                result = "heads"
                print("The coin landed on heads")
            else:
                result = "tails"
                print("The coin landed on tails")
                
            #Compare results    
            if result == coin_flip_guess:
                print("You guessed correct!")
                correct_guesses += 1
                total_guesses += 1
            else:
                print("Sorry, you guessed wrong.")
                total_guesses += 1
            
            print("You have correctly guessed " + str(correct_guesses) + " out of " + str(total_guesses))
            
            #Ask to play again
            play_again = input ('Do you want to play again Y/N: ')
            if play_again == 'Y':
                index = 0
            else:
                index = 1
    print("Thanks for playing!")
   
```
